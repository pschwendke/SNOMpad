```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv
from scipy.optimize import root
```

```python
def bessel_difference(gamma, m):
    return jv(m, gamma) - jv(m+1, gamma)
```

```python
x = np.arange(0, 10, 0.01)
y1 = bessel_difference(x, 1)
y2 = bessel_difference(x, 2)
y3 = bessel_difference(x, 3)
ys = [y1, y2, y3]

for y in ys: plt.plot(x, y)
plt.show()
```

```python
gamma1 = float(root(bessel_difference, 2.5, args=1).x)
gamma2 = float(root(bessel_difference, 4, args=2).x)
gamma3 = float(root(bessel_difference, 5, args=3).x)
gammas = [gamma1, gamma2, gamma3]
```

```python
for i in range(3):
    plt.plot(x, ys[i], label=fr'm = {i+1},  $\gamma$ = ' + str(gammas[i]))
    plt.axvline(gammas[i], c='black', lw=1, ls='--')

plt.axhline(0, c='gray', lw=1)

plt.legend()
plt.show()
```
